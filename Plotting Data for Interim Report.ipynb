```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

crime= pd.read_csv(r"C:\Users\Adnan\MIS Project\Crime-Analysis-master\datasets\generated.csv")

# Plotting regions against count
crime_region=crime['region']
crime_count=crime['count']
plt.bar(crime_region,crime_count,align='center')
plt.xlabel('Regions of NY')
plt.ylabel('Total crimes committed')
plt.title("Crimes committed in different counties")
```

```python
# Plotting crime against years
crime_year=crime['year']
plt.bar(crime_year,crime_count,align='center')
plt.xlabel('Years')
plt.ylabel('Total crimes committed')
plt.xticks(range(1990,2020, 3))
plt.title("Crime committed since 1990")
```

```python
# Sorting ouot Bronx from list
br = ['BRONX']
bronx=crime[crime.region.isin(br)]

bronx_count=bronx['count']
bronx_year=bronx['year']

# Plotting with Bronx crimes against years
plt.xlabel('Years')
plt.ylabel('Total crimes committed')
plt.bar(bronx_year,bronx_count,align='center')
plt.xticks(range(1990,2020, 5))
plt.title("Bronx")
```
